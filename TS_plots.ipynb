# Time series plot with marked EVENT

```python
df['label'] = np.where(df['EVENT']==True, 1, -1)

  # plot
# =============================
#df = df.dropna(axis=0, how='any')
#fig, ax = plt.subplots(8, sharex =True ,figsize=(15,24))
fig, ax = plt.subplots(figsize=(6.5,1.33))
#plt.xlabel('Time')
plt.rcParams['xtick.labelsize'] = 7
def plot_func(group):
    global ax
    color = 'b' if (group['label'] < 0).all() else 'r'
    lw = 0.5
    ax.set_title('Temperature')
    ax.plot(group.index, group.Tp, c=color, linewidth=lw)
    #ax[1].plot(group.index, group.Cl, c=color, linewidth=lw)
    #ax[2].plot(group.index, group.pH, c=color, linewidth=lw)
    #ax[3].plot(group.index, group.Redox, c=color, linewidth=lw)
    #ax[4].plot(group.index, group.Leit, c=color, linewidth=lw)
    #ax[5].plot(group.index, group.Cl_2, c=color, linewidth=lw)
    #ax[6].plot(group.index, group.Fm, c=color, linewidth=lw)
    #ax[7].plot(group.index, group.Fm_2, c=color, linewidth=lw)
df.groupby((df['label'].shift() * df['label'] < 0).cumsum()).apply(plot_func)
plt.savefig('Images/temperature.png', dpi=500, bbox_inches='tight')
```
