```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Distribution of training data based event

```python
def groupby_event(data):
    data_grouped_by_event = data.groupby(data.EVENT)
    labels = data.EVENT.unique()
    counts = []
    for label in labels:
        counts.append(data_grouped_by_event.get_group(label).count().Trueb.astype(str))
    return labels, counts
```

```python
training_data = pd.read_csv('./Data/training2018.csv', header=0, index_col=0)
dataset = pd.read_csv('./Data/imputedWQ.csv', header=0, index_col=0)
```

```python
plt.figure(1, figsize=(8,6))
# Data to plot
types, counts = groupby_event(training_data)
colors = ['lightcoral', 'lightgreen']
#explode = (0, 0.1, 0, 0, 0)  # explode largest slice
 
# Plot
plt.pie(counts, labels=types, autopct='%1.1f%%', startangle=90, colors=colors)
plt.legend(fancybox=True)
plt.axis('equal')
plt.rc('text', usetex=True)
plt.savefig('training_distribution.png', transparent=True)
plt.show()
```

```python
types, sizes = groupby_event(training_data)

counts = [float(i) for i in sizes]
percentage = [round(count/sum(counts)*100, 1) for count in counts]
#percentage = map(lambda x: round(x*100, 1), percentage)
```

```python
plt.rc('text', usetex=False)
plt.figure(1, figsize=(2.31,1.75))

y_pos = np.arange(len(types))
fig, ax = plt.subplots()
plt.barh(y_pos, percentage,height=0.4, align='center', alpha=0.8, color='steelblue')
plt.yticks(y_pos, types)
plt.xticks(range(0,50,10))
plt.xlim([0, 100])
 

plt.xlabel('Percentage(%)')
#plt.ylabel('Transportation Mode')
plt.title('Training data distribution')

plt.grid(True, axis='x', linestyle='--')
plt.axvline(x=20, linewidth=1, color='steelblue', linestyle='--')

plt.rc('font', size=20)
#plt.rc('font', family='normal') 
plt.rcParams['font.size'] = 20
plt.rcParams['axes.labelsize'] = 20
#plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 18
plt.rcParams['figure.titlesize'] = 20
#plt.savefig('images/training_distribution_barh.png', dpi=500, bbox_inches='tight')
plt.show()
```

```python
from sklearn import datasets
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Loading dataset
dataset = pd.read_csv('./Data/imputedWQ.csv', header=0, index_col=0)
```

```python
dataset_wo_event = dataset.loc[:, dataset.columns != 'EVENT']
```

```python
# Defining Model
model = TSNE(learning_rate=100)
```

```python
# Fitting Model
transformed = model.fit_transform(dataset_wo_event)
```

```python
# Plotting 2d t-Sne
x_axis = transformed[:, 0]
y_axis = transformed[:, 1]
```

```python
plt.scatter(x_axis, y_axis, c=dataset.EVENT)
plt.show()
```
