```python
# line plot of time series
from pandas import Series
from matplotlib import pyplot
# load dataset
series = Series.from_csv('TSdata/car-sales1.csv', header=0)
# display first few rows
print(series.head(5))
```

```python
# line plot of dataset
#pyplot.xlabel('EVENT', fontsize=4)
pyplot.figure(1, figsize=(2.95,2.2))
pyplot.ylabel('Car Sales', fontsize=10)
series.plot(color='black')
pyplot.rcParams['xtick.labelsize'] = 7
pyplot.rcParams['ytick.labelsize'] = 7
pyplot.savefig('img1.png', dpi=500, bbox_inches='tight')
pyplot.show()
```
