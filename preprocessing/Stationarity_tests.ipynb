# ADF test

df1- Training dataset
df2- Testing dataset
df- df1 and df1 combined

From the ADF test we have:
In df1 only the Temperature is non-stationary, all other variables are stationary.
In df2 , we have stationarity on most of the variables, but some are not significat(pH,Redox), Temperture here is also stationary with a p-value= 0.02 .
In df dataset- Temperature is non-stationary, other varialbles are stationary.

```python
from statsmodels.tsa.stattools import adfuller, kpss
```

```python
Tp.head()
```

```python
def adf_test(y):
    # perform Augmented Dickey Fuller test
    print('Results of Augmented Dickey-Fuller test:')
    dftest = adfuller(y, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['test statistic', 'p-value', '# of lags', '# of observations'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value ({})'.format(key)] = value
    print(dfoutput)
```
